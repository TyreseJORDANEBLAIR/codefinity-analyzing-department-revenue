## Basic Filtering

```python
import pandas as pd

# 1. Load both datasets
departments = pd.read_csv('departments_structure.csv')
performance = pd.read_csv('departments_performance.csv')

# 2. (Optional) Quick inspection
print(departments.head())
print(performance.head())

# 3. Clean (if needed)
departments = departments.dropna().drop_duplicates()
performance = performance.dropna().drop_duplicates()

# 4. Merge on DEPARTMENT_ID
merged_df = pd.merge(departments,
                     performance,
                     on='DEPARTMENT_ID')

# 5. Filter for REVENUE > 3 000 000 AND LOCATION_ID is 1800 or 1400
common_departments = merged_df[
    (merged_df['REVENUE'] > 3000000) 
    & (merged_df['LOCATION_ID'].isin([1800, 1400]))
]

# 6. Reset index
common_departments = common_departments.reset_index(drop=True)

# 7. Display result
print(common_departments)
```

## Merging the Data

## Filtering Data with Multiple Conditions

```python
import pandas as pd 

departments = pd.read_csv('departments_structure.csv')
performance = pd.read_csv('departments_performance.csv')

merged_df = pd.merge(departments, performance, on='DEPARTMENT_ID')

successful_departments = merged_df[
    (merged_df['REVENUE'] > 3000000)
    & (merged_df['LOCATION_ID'].isin([1800, 1400]))
]

successful_departments = successful_departments.reset_index(drop=True)

print(successful_departments)
```

## Grouping the Data

```python
import pandas as pd 

departments = pd.read_csv('departments_structure.csv')
performance = pd.read_csv('departments_performance.csv')

merged_df = pd.merge(departments, performance, on='DEPARTMENT_ID')

successful_departments = merged_df[
    (merged_df['REVENUE'] > 3000000)
    & (merged_df['LOCATION_ID'].isin([1800, 1400]))
]

successful_departments = successful_departments.reset_index(drop=True)

print(successful_departments)
```

```python
# Group by LOCATION_ID: sum of REVENUE and average of EMPLOYEE_COUNT
grouped_stats = successful_departments.groupby('LOCATION_ID').agg(
    REVENUE_SUM=('REVENUE', 'sum'),
    EMPLOYEE_COUNT_AVG=('EMPLOYEE_COUNT', 'mean')
).reset_index()

print(grouped_stats)
```
